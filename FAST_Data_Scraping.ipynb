```python
import requests
import re
from bs4 import BeautifulSoup
from IPython.display import display
import pandas as pd

URL="https://lhr.nu.edu.pk/faculty/"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")

alldepts = ['cv', 'fsc', 'ee', 'fsm', 'ss']
allpd=[]

for i in alldepts:
    department = soup.find('div', id=i)
    education_link = department.find_all(href=True)
    education_link1 = ['https://lhr.nu.edu.pk/'+x['href'] for x in education_link]

    education=[]
    for x in education_link1:
        page1 = requests.get(x)
        s = BeautifulSoup(page1.content, "html.parser")
        hr_tag = s.find('hr', class_='nu-hr')
        ultag=hr_tag.find_next("ul")
        x = ultag.find_all("li")
        temp = [a.text for a in x]
        education.append(', '.join(temp))

    img = ['https://lhr.nu.edu.pk/'+x['src'] for x in department.findAll('img', class_='card-img-top')]
    first_h1 = department.find('h1')
    allnames = [x.text for x in department.findAll('h5', class_='text-center')]
    email = [x.text for x in department.findAll('p', class_='mb-0 text-center')]
    design = [x.text for x in department.findAll('p', class_='small text-center font-italic')]
    temp = [x.text.replace('\n', '').strip() for x in department.findAll('p', class_='small text-center font-italic')]
    cleaneddesign= [re.sub(r'\s+', ' ', item).strip() for item in temp]
    highest = [allnames[x] if "HEC" in cleaneddesign[x] else "" for x in range(len(cleaneddesign))]

    id = [education_link1[x].split('/')[-1] for x in range(len(education_link1))]
    
    allpd.append(pd.DataFrame({'ID':id, 'Name': allnames, 'Email': email, 'Designation': cleaneddesign, 'HEC Approved PhD': highest ,'Department':first_h1, 'Education':education, 'Extension':id,'image URL': img}))
```

```python
print("Lahore Campus")
for x in range(len(allpd)):
    if x == 0:
        print("School of Civil Engineering")
    elif x == 1:
        print("FAST School Of Computing")
    elif x == 2:
        print("School of Electrical Engineering")
    elif x == 3:
        print("FAST School Of Management")
    elif x == 4:
        print("FAST School Of Social Science")
    display(allpd[x])
```

```python
import requests
import re
from bs4 import BeautifulSoup
import pandas as pd

URL = "https://khi.nu.edu.pk/"
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")

faclink = soup.find('li', class_='menu-item-6256')
allpdkh = []

if faclink:
    atags = faclink.find_all('a')

    for x in range(len(atags)):
        href = atags[x]['href']
        department = atags[x].text

        if href == '#':
            continue

        p1 = requests.get(href)
        s1 = BeautifulSoup(p1.content, "html.parser")

        f1 = s1.find_all('div', class_='gdlr-core-personnel-list-image')
        allnameskh = []
        emailkh = []
        designkh = []
        imglink = []
        education = []
        idget = []
        dept = []
        hecapprovedkh = []

        for element in f1:
            atag = element.find('a')
            img = element.find('img')

            allnameskh.append(img['title'])
            p2 = requests.get(atag['href'])
            s2 = BeautifulSoup(p2.content, "html.parser")
            design = s2.find_all('span', class_='gdlr-core-skin-caption')
            ed1 = s2.find_all('i', class_='fa-graduation-cap')
            email = s2.find('div', class_='kingster-type-email')
            pattern = re.compile(r'"page_id":"(\d+)"')
            matches = pattern.search(str(s2))

            if matches:
                page_id = matches.group(1)
                idget.append(page_id if page_id else '')

            if email:
                emailkh.append(email.text)
            else:
                emailkh.append('')

            for element in ed1:
                parent = element.find_parent()
                sibling = parent.find_next_sibling()
                if sibling:
                    education.append(sibling.text)
                else:
                    education.append('')

            design1 = design[0].text
            if "HEC" in design1:
                hecapprovedkh.append('HEC Approved')
            else:
                hecapprovedkh.append('')

            designkh.append(design1)
            imglink.append(img['src'])
            dept.append(department)

        max_length = max(len(idget), len(allnameskh), len(emailkh), len(designkh), len(hecapprovedkh), len(dept), len(education), len(imglink))
        idget.extend([''] * (max_length - len(idget)))
        allnameskh.extend([''] * (max_length - len(allnameskh)))
        emailkh.extend([''] * (max_length - len(emailkh)))
        designkh.extend([''] * (max_length - len(designkh)))
        hecapprovedkh.extend([''] * (max_length - len(hecapprovedkh)))
        dept.extend([''] * (max_length - len(dept)))
        education.extend([''] * (max_length - len(education)))
        imglink.extend([''] * (max_length - len(imglink)))

        allpdkh.append(pd.DataFrame({'ID': idget, 'Name': allnameskh, 'Email': emailkh, 'Designation': designkh, 'HEC Approved PhD': hecapprovedkh, 'Department': dept, 'Education': education, 'Extension': idget, 'image URL': imglink}))
else:
    print("No matching <li> element found.")

final_df = pd.concat(allpdkh, ignore_index=True)

print(final_df)
```
